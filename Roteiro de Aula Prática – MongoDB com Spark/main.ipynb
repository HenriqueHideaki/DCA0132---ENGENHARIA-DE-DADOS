```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("StreamingApp").getOrCreate()
```

```python
meuschema = "ID_PROC int, VARA int, CLASSE int, QTD_RDA_PROC int, QTD_RTE_PROC int, ID_FAIXA int, ID_RDA int, QTD_ASSUNTOS int, FAIXA_CAUSA string, TEMPO_DIAS int, ASSUNTO string"
```

```python
dfEntrada = spark.readStream.format("csv").option("header", "true").schema(meuschema).load("discentes-2022.csv")
```

```python
dfSaida = dfEntrada.filter("TEMPO_DIAS > 3000")
dfSaida
```

```python
dfSaida.writeStream \
    .format("console") \
    .outputMode("append") \
    .trigger(processingTime="20 seconds") \
    .option("checkpointlocation", "C:\Hidea\spark\pasta_checkpoint") \
    .start()
```
